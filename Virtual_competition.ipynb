**Diabetes dataset EDA**

````python
data = pd.read_csv('/content/diabetes_model.csv')
data
````

UNIVARIATE ANALYSIS

**using histograms**

````python
import seaborn as sns
sns.histplot(data['Glucose'],bins=10,color='powderblue',edgecolor='white')
plt.show()
````

The above graphs shows a histogram representation with Glucose in its x axis and count in y axis. The graph is right skewed.The majority of the glucose level falls between 100 to 125 range.

````python
sns.histplot(data['BloodPressure'],bins=10,color='lightcoral',edgecolor='white')
plt.show()
````

The graph is a histogram representation with Bloodpressure in its x axis and count in its y axis. The count for high Bp is comparitevely with lesser count
ranges inbetween 0 to 25 . The maximum count shows Bp ranging from 60 to 80.


**using boxplots**

````python
sns.boxplot(data['Pregnancies'],color='steelblue',width=0.3)
plt.show()
````

The above given boxplot gives a median of range 5.5 approximately.The box shows the range of pregnancies which lies between the interquartile range.

**BIVARIATE ANALYSIS**

Using correlation matrix

````python
corr_matrix= data[['Pregnancies','BloodPressure']]
corr_matrix
corr = corr_matrix.corr()
plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap='cividis')
plt.show()
````

The above shows a correlation matrix  defining the variables BloodPressure and Pregnancies.The 1.0 range show positive relation , and the negative relation ranges below 1.0 and weak relation.The weak relation can either increase or decrease together.

**MULTIVARIATE ANALYSIS**

````python

sns.pairplot(data[['Glucose','BloodPressure','BMI','Age']],hue='Glucose',palette='mako')
plt.show()
````

The plot shows pairplot with variables of age ,bmi and bloodpressure.The diagonal plots display the distribution of each variable.The scatter plots show how these variables are correlated. For example, there seems to be a positive correlation between BMI and BloodPressure, meaning that as BMI increases, BloodPressure tends to increase as well.
