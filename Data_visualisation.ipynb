```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

Violin plot

```python
plt.figure(figsize=(6, 6))
sns.violinplot(x='species', y='petal_length', data=iris,hue='species', palette='pastel')
plt.title('Violin Plot of Petal Length by Species')
plt.show()


#Violin plot provides a detailed view of the distribution of petal lengths across different iris species.
```

Swarmplot

```python
sns.swarmplot(x='species', y='sepal_width', data=iris,hue='petal_width', palette='pastel')
plt.title('Swarm Plot of Sepal width by Species')
plt.show()

#visualize the distribution of sepal width across different iris species, with an additional dimension represented by petal_width'hue' parameter colors the points based on the petal_width variable.
```

Stripplot

```python
plt.figure(figsize=(6, 4))
sns.stripplot(x='species', y='sepal_length', data=iris, jitter=True,hue='species', palette='Set2')   # jitter helps to make overlapping points more visible
plt.title('Strip Plot of Sepal Length by Species')
plt.show()
```

Boxplot

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='species', y='petal_width', data=iris,color="powderblue")
plt.title('Box Plot of Petal Width by Species')
plt.show()
```

Barplot


```python
sns.barplot(x='species', y='sepal_length', data=iris,palette="flare",width=0.35,edgecolor='white',hue = 'species')
plt.title('Bar Plot of Sepal Length by Species')
plt.show()
```

Countplot

```python
sns.countplot(x='species', data=iris,hue='species', palette='crest')
plt.title('Count Plot of Iris Species')
plt.show()
```

Heatmap

```python
iris_numeric = iris[['sepal_length','sepal_width','petal_length','petal_width']]
corr = iris_numeric.corr()
plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.show()

# 1 indicates a perfect positive correlation.
#-1 indicates a perfect negative correlation.
#0 indicates no linear correlation
```

Scatterplot

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris, palette='mako_r')
plt.title('Sepal Length  vs  Sepal Width')
plt.show()


#The scatter plot visualizes how sepal length and sepal width are related for each iris flower.
#The 'hue' parameter colors the points based on the species of the iris flowers.
#color palette used for different species is 'mako_r'
```

Lineplot

```python
sns.lineplot(x='species', y='sepal_width',data=iris, marker='*',markersize=18,color="Purple")
plt.title('Line Chart of Average Sepal Length by Species')
plt.show()
```
